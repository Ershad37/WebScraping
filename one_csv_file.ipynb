# Read one csv file from URL with pandas

### Target website: https://www.football-data.co.uk/data.php

```python
# reading the dataset using read_csv method
df_bundesliga = pd.read_csv("https://www.football-data.co.uk/mmz4281/2425/D1.csv")
```

```python
# show the dataframe
df_bundesliga
```

```python
# rename some columns to understand better the dataframe
df_bundesliga.rename(columns={"FTHG": "home_goals",
                              "FTAG": "away_goals"}, inplace=True)
```

```python
# we use columns attribute to get the names of columns
df_bundesliga.columns
```
